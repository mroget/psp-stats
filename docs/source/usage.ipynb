# Usage

This notebook demonstrates how to use the `qpsp_stats` package.

```python
# Imports
import qpsp_stats as qpsp # The package
```

## 1. Sampling SAW
One can sample Self Avoiding Walk (SAW) either at random (via Pivot or dimerize algorithm) or by enumerating everything.

```python
# At random using the Pivot
l = qpsp.sample_solutions(len=100, sample_size=10000, method="pivot", lattice="tetrahedral")
print("Number of samples:", len(l)) # print 10000
print("One solution:\n", l[0][:5]) # First five coordinates of the first random SAW
```

```python
# Enumerating all SAW
l = qpsp.sample_solutions(len=13, method="enumerate", lattice="tetrahedral")
print("Number of samples:", len(l)) # print 149499. This number increases exponentially in len !
print("One solution:\n", l[0][:5]) # First five coordinates of the first random SAW
```

## 2. Calculating statistics for a cost function and a protein

```python
import prody # Loading pdb files
import numpy as np
```

```python
protein = prody.parsePDB("2k2r")["A"] # Loading PDB file
sequence = protein.select("name CA").getSequence() # Extracting the sequence
ground_truth = np.array(protein.select("name CA").getCoordsets())  # Extracting the coordinates of the CA atoms
print(f"Protein chain of length {len(sequence)} loaded !")
```

Now we can calculate the correlation between the rmsd against ground truth and the cost function. A length 129 has too many possible SAW for enumerating them all, so we use random sampling instead.

```python
corr = qpsp.correlation(sequence, ground_truth, lambda seq,coords:np.random.random(), sample_size=1000, method="pivot", lattice="tetrahedral")
print("Correlation:", corr)
```

Cell above returns a correlation coefficient close to 0. That is because the cost function we provided returns random numbers without reading the input. They are not correlated, hence a coefficient close to 0.

$$
cost(\text{seq}, \text{structure}, MJ) = \sum_{0 \leq i < j < \text{len}(\text{seq})} \epsilon_{i,j}(\text{seq},\text{structure},MJ),
$$
where 
$$
\epsilon_{i,j}(\text{seq},\text{structure},w) = \left\{
	\begin{matrix}
		w_{\text{seq}_i,\text{seq}_j} & \text{if } \Vert \text{structure}_i - \text{structure}_j \Vert_2 \leq 7.8\\
		0 & \text{otherwise}\\
	\end{matrix}
		\right. .
$$
